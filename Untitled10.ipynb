```python
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.efficientnet import preprocess_input, decode_predictions
```

подготавливаем даталоудеры формат папок
# dataset/
# ├── cats/
# │   ├── cat1.jpg
# │   └── cat2.jpg
# └── dogs/
#     ├── dog1.jpg
#     └── dog2.jpg
можем писать rescale=1./255 вместо preprocessing_function=preprocess_input но только для реснет


```python
train_gen = ImageDataGenerator(preprocessing_function=preprocess_input, validation_split=0.2).flow_from_directory('/content/drive/MyDrive/archive/train', target_size=(224,224), batch_size=32, subset='training')
val_gen = ImageDataGenerator(preprocessing_function=preprocess_input, validation_split=0.2).flow_from_directory('/content/drive/MyDrive/archive/test', target_size=(224,224), batch_size=32, subset='validation')
```

Загружаем модель нейронка рекомендует EfficientNetB0

```python
base = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224,224,3))
base.trainable = False
```

```python
model = tf.keras.Sequential([base, GlobalAveragePooling2D(), Dense(train_gen.num_classes, activation='softmax')])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_gen, validation_data=val_gen, epochs=10)
```

```python
labels_dict = train_gen.class_indices

# Получить просто список названий в правильном порядке
class_names = list(labels_dict.keys())

print(class_names)
```

Проверка на изображении

```python
img = tf.keras.utils.load_img('/content/21.jfif', target_size=(224, 224))
img_array = tf.keras.utils.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)
```

```python
predictions = model.predict(img_array, verbose=0)[0]
class_idx = np.argmax(predictions)
confidence = predictions[class_idx]
```

Нет, обучать с нуля почти никогда не нужно, даже если ваши изображения сильно отличаются от ImageNet (например, это медицинские рентгены, спутниковые снимки или микросхемы).Вот почему Transfer Learning (перенос обучения) всё равно работает:Универсальные признаки: Первые слои любой нейросети учатся видеть простые вещи: линии, углы, цветовые пятна и текстуры. Эти «кирпичики» одинаковы и в фотографиях котиков из ImageNet, и в снимках МРТ.Экономия времени и данных: Обучение с нуля требует в 10–100 раз больше данных и на порядок больше времени. Если у вас меньше 50 000–100 000 картинок, обучение с нуля, скорее всего, даст результат хуже, чем дообучение готовой модели.Как действовать, если изображения специфические:Вместо обучения с нуля используйте стратегию глубокого Fine-tuning:Этап 1 (Разминка): Замораживаете base (как вы делали сейчас) и обучаете только «голову» (Dense слой) 3–5 эпох. Это нужно, чтобы случайные веса нового классификатора не «испортили» градиентами предобученную базу.Этап 2 (Разморозка): Размораживаете всю модель (base.trainable = True) и обучаете на очень маленьком Learning Rate (например, 1e-5 или 1e-6).Когда ВСЁ-ТАКИ нужно обучение с нуля?Только в экстремальных случаях:Если ваши данные — это не обычные 3-канальные фото (например, 15-канальные спутниковые данные или спектрограммы).Если у вас огромный датасет (миллионы изображений) и специфическая задача, где веса ImageNet могут «сбивать» модель с толку.Что именно на ваших изображениях? (Например: медицина, чертежи, ночная съемка или что-то другое?)
